```python
import tensorflow as tf

# X and Y data
x_train = [1,2,3]
y_train = [1,2,3]

W = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

#Our hypothesis XW+B
hypothesis = x_train * W + b
#cost/loss finction
cost = tf.reduce_mean(tf.square(hypothesis-y_train))

#minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost) #optimizer를 통해서 찾은 cost의 최솟값을 train에 저장한다

```

```python
#lauch the graph in a session
sess = tf.Session()
#initializes global variables in the graph
#Variable을 사용하는 경우에 반드시 Varialbe들을 초기화해주어야한다.
sess.run(tf.global_variables_initializer())

#Fit the line
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

```python
#placeholder를 이용한 graph
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

W = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis2 = X * W + b
cost = tf.reduce_mean(tf.square(hypothesis2-Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)


#launch
sess.run(tf.global_variables_initializer())

#fit the lin
for step in range(2001):
    cost_val, W_val, b_val, _ = \
        sess.run([cost,W,b,train],
                 feed_dict={X: [1,2,3], Y:[1,2,3]})
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
    
```
